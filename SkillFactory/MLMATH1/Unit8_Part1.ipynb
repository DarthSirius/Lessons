# 8. Практика: градиентный спуск

```python
def f(x, y):
    return 2*x**2 - 4*x*y + y**4 + 2
def grad(x, y):
    dx = 4*x - 4*y
    dy = -4*x + 4*y**3
    return (dx, dy)
```

```python
x0 = (0.5, 0.3)
gamma = 0.01
x_cur = x0

vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1])
    if i > 1000:
        break
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.4f}, {x_cur[1]:.4f});"
          f" f(x)={f(*x_cur):.4f}; grad f(x)=({grad(*x_cur)[0]:.4f}, {grad(*x_cur)[1]:.4f})")
```

Найдите минимум функции
с помощью градиентного спуска.

```python
def f(x, y, z):
    return x**3 - 2*x**2 + y**2 + z**2 - 2*x*y + x*z - y*z + 3*z
def grad(x, y, z):
    dx = 3*x**2 - 4*x - 2*y + z
    dy = 2*y - 2*x - z
    dz = 2*z + x - y + 3
    return (dx, dy, dz)
```

```python
x0 = (0.5, 0.3, 0.2)
gamma = 0.01
x_cur = x0

vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1],
            x_cur[2] - gamma * grad(*x_cur)[2])
    if i > 1000:
        break
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.4f}, {x_cur[1]:.4f});"
          f" f(x)={f(*x_cur):.4f}; grad f(x)=({grad(*x_cur)[0]:.4f}, {grad(*x_cur)[1]:.4f})")
```

```python
optimize.minimize(lambda x: f(*x), x0=(0, 0)) 
```
