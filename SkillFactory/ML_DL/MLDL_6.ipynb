```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
houses_data = pd.read_csv('data/boston.csv')
houses_data
```

```python
columns = ['crim_rate','business','nit_oxiden','zn','rooms','age','dist','pup_per_teaс','lower','target']
for column in columns:
    houses_data[column] = houses_data[column].apply(lambda x: x.replace(',','.'))
    houses_data[column] = pd.to_numeric(houses_data[column])
    
houses_data
```

```python
houses_data.info()
```

```python
X = houses_data.drop(['target'], axis=1).values
y = houses_data['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, shuffle=False)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(mean_squared_error(y_test, y_pred))
```

Заведите массив для объектов DecisionTreeRegressor (будем их использовать в качестве базовых алгоритмов) и для вещественных чисел (это будут коэффициенты перед базовыми алгоритмами).

```python
def gbm_predict(X):
    return [sum([coeff * algo.predict([x])[0] for algo, coeff in zip(base_algorithms_list, coefficients_list)]) for x in X]
```

```python
coef = 0.9
tree = DecisionTreeRegressor(max_depth=5, random_state=139)

base_algorithms_list = [tree.fit(X_train, y_train)]
coefficients_list = [coef]

for i in range(1,50):
    y_pred = gbm_predict(X_train)
    base_algorithms_list.append(tree.fit(X_train, y_train - y_pred))
    coefficients_list.append(coef)
    mse = mean_squared_error(y_train, y_pred)
    print(i, mse)

y_pred = gbm_predict(X_test)
np.sqrt(mean_squared_error(y_test, y_pred))
```
