# Базовые возможности библиотеки Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('default')
```

1. Line praph / Линейный график

```python
year = [1950, 1975, 2000, 2018]
population = [2.12, 3.681, 5.312, 6.981]
plt.plot(year, population)
plt.show()
```

2. Scatterplot Диаграмма рассеяния

```python
plt.scatter(year, population)
```

```python
values = [0, 1.2, 1.3, 1.9, 4.3, 2.5, 2.7, 4.3, 1.3, 3.9] 
plt.hist(values, bins = 4) 
plt.show()
```

```python
import numpy as np
X = np.linspace(- np.pi, 2 * np.pi) 
cos, sin = np.cos(X), np.sin(X) 
plt.plot(X, cos, color='blue', label="cosine") 
plt.plot(X, sin, color='red', label="sine") 
plt.legend(loc='upper left', frameon=False)
plt.show()
```

```python
# Data to plot 
names = 'Tom', 'Dick', 'Harry', 'Jill', 'Meredith', 'George' 
speed = [8, 7, 12, 4, 3, 4] 
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red', 'blue'] 
explode = (0.1, 0, 0, 0, 0, 0) 
# explode 1st slice 
plt.pie(speed, explode=explode, labels=names, colors=colors, autopct='%1.1f%%', shadow=False, startangle=90) 
plt.show()
```

```python
# Create data 
temperature = np.random.randn(4096) 
anger = np.random.randn(4096) 
# Create heatmap 
heatmap, xedges, yedges = np.histogram2d(temperature, anger, bins=(64,64)) 
# extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]] 
# Plot heatmap 
# plt.clf() 
plt.ylabel('Anger')
plt.xlabel('Temp') 
plt.imshow(heatmap) 
plt.show()
```
