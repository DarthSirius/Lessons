### WEB

```python
import requests
url = 'https://www.cbr-xml-daily.ru/daily_json.js'
response = requests.get(url)
```

```python
print(response.text)
```

```python
from pprint import pprint # Импортируем функцию pprint()
currencies = response.json() # Применяем метод json()
pprint(currencies)
```

```python
pprint(currencies['Valute']['EUR'])
```
